# compress vocab

```python
from tqdm import tqdm_notebook as log_progress


def load_lines(path):
    with open(path) as file:
        for line in file:
            yield line.rstrip()


def load_vocab(path):
    lines = load_lines(path)
    for line in lines:
        word, count = line.split(None, 1)
        yield word, int(count)
        
        
def vocab_path(name):
    return 'vocabs/%s.txt' % name


# ['librusec', 'taiga_proza', 'news', 'wiki']
path = vocab_path('librusec')
records = load_vocab(path)
words, counts = [], []
for word, count in log_progress(records):
    words.append(word)
    counts.append(count)
pairs = list(zip(words, counts))
```

```python
import gzip

text = '\n'.join(words)
bytes = text.encode('utf8')
data = gzip.compress(bytes)
%time gzip.decompress(data)
len(data) / 1024 / 1024
```

```python
import numpy as np
x = np.array(counts)
y = x.astype(np.uint32)
data = gzip.compress(y.tobytes())
%time np.frombuffer(gzip.decompress(data))
len(data) / 1024 / 1024
```

```python
import numpy as np
x = np.array(counts)
y = x.astype(np.uint32)
data = y.tobytes()
%time np.frombuffer(data)
len(data) / 1024 / 1024
```

```python
from io import BytesIO

buffer = BytesIO()
file = gzip.GzipFile(mode='wb', fileobj=buffer)
text = '\n'.join(words)
bytes = text.encode('utf8')
file.write(bytes)
file.flush()
file.close()
data = buffer.getvalue()
```

# vocab

```python
from collections import OrderedDict

def load_lines(path):
    with open(path) as file:
        for line in file:
            yield line.rstrip()


def load_vocab(path):
    lines = load_lines(path)
    for line in lines:
        word, _ = line.split(None, 1)
        yield word
        
        
def vocab_path(name):
    return 'vocabs/%s.txt' % name


vocabs = OrderedDict()
for name in ['librusec', 'taiga_proza', 'news', 'wiki']:
    words = set(load_vocab(vocab_path(name)))
    vocabs[name] = words
```

```python
for a in vocabs:
    for b in vocabs:
        if a < b:
            print(len(vocabs[a] & vocabs[b]), len(vocabs[a]), len(vocabs[b]), len(vocabs[a] & vocabs[b]) / min(len(vocabs[a]), len(vocabs[b])), a, b, sep='\t')
```

```python
len({
    word
    for name in vocabs
    for word in vocabs[name]
})
```

```python
from collections import Counter
from os.path import expanduser, join
from os import listdir

from nerus.load import load_raw
from navec.train.tokens import find_tokens


dir = expanduser('~/proj/nerus/data/dumps')
freqs = OrderedDict()
for filename in ['factru.raw.jsonl.gz', 'gareev.raw.jsonl.gz', 'ne5.raw.jsonl.gz']:
    print(filename)
    path = join(dir, filename)
    records = load_raw(path)
    counts = Counter()
    for record in records:
        text = record.source.text
        for span in record.source.spans:
            for word in find_tokens(text[span.start:span.stop]):
                counts[word.lower()] += 1
#         for word in find_tokens(record.source.text):
#             counts[word.lower()] += 1
    freqs[filename] = counts
```

```python
def get_cover(counts, vocab):
    total = 0
    in_vocab = 0
    for word, count in counts.items():
        total += count
        if word in vocab:
            in_vocab += count
    return in_vocab / total


for filename in freqs:
    for vocab in vocabs:
        print(get_cover(freqs[filename], vocabs[vocab]), filename, vocab)
```

```python
from tqdm import tqdm_notebook as log_progress
from collections import Counter
from navec.train.glove import load_glove_vocab, load_glove_cooc

counts = Counter()
words = [word.decode('utf8') for word, _ in load_glove_vocab('vocabs/all.txt')]
records = load_glove_cooc('t')
for (source, target), weight in log_progress(records):
    if source != target:
        counts[words[source - 1], words[target - 1]] = weight
```

```python
from navec import Navec

navec = Navec.load('../data/models/navec/navec_hudlit_v1_12B_500K_300d_100q.tar')
```

```python
kv = navec.as_gensim
```

```python
def analogy(a, b, c):
    return kv.most_similar(positive=[c, b], negative=[a])
```
